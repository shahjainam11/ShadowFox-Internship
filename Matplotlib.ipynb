```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.show()
```

```python

categories = ['A', 'B', 'C']
values = [10, 15, 7]
plt.bar(categories, values)
plt.title('Bar Chart')
plt.show()
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
plt.hist(data, bins=5)
plt.title('Histogram')
plt.show()
```

```python
sizes = [30, 45, 25]
labels =['Java', 'Python', 'C++']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
data = [7, 15, 13, 20, 9, 18, 22, 5, 11, 14]
plt.boxplot(data)
plt.title('Box Plot')
plt.ylabel('Values')
plt.show()
```
