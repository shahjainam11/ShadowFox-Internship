```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x=[1, 2, 3], y=[5, 12, 15])
plt.title('Line Plot')
plt.show()
```

```python
sns.scatterplot(x=[1, 2, 3], y=[4, 5, 6])
plt.title('Seaborn Scatter Plot')
plt.show()
```

```python
import pandas as pd
data = {'group': ['A', 'B', 'C'], 'score': [5, 7, 4]}
df = pd.DataFrame(data)
sns.barplot(x='group', y='score', data=df)
plt.title('Bar Plot')
plt.show()
```

```python
data = [1, 2, 2, 3, 4, 4, 5]
sns.histplot(data, kde=False)
plt.title('Histogram')
plt.show()
```

```python
data = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=data)
plt.title('Box Plot')
plt.show()
```

```python
data = sns.load_dataset("tips")
day_counts = data['day'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(day_counts, labels=day_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.axis('equal')
plt.show()
```
